```python
import requests
response = requests.get('https://api.github.com/events')

if response.status_code == 200:
    print("Success!")
    print(response.url) # This will print the URL of the request
    print(response.headers) # This will print the headers of the response
    print(response.headers['Content-Type']) # This will print the content type of the response
else:
    print("Failed to retrieve data. Status code:", response.status_code)
   
```

```python
print(response.text)  # This will print the content of the response as a string
```

```python
data = response.json()  # This will parse the JSON response into a Python dictionary
print(data)  # This will print the parsed JSON data
print(type(data))  # This will print the type of the parsed data
```
