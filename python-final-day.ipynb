```python
import requests
response = requests.get('https://api.github.com/events')

if response.status_code == 200:
    print("Success!")
    print(response.url) # This will print the URL of the request
    print(response.headers) # This will print the headers of the response
    print(response.headers['Content-Type']) # This will print the content type of the response
else:
    print("Failed to retrieve data. Status code:", response.status_code)
   
```

```python
print(response.text)  # This will print the content of the response as a string
```

```python
data = response.json()  # This will parse the JSON response into a Python dictionary
print(data)  # This will print the parsed JSON data
print(type(data))  # This will print the type of the parsed data
```

```python
import json
import requests
url = 'https://api.github.com/events'
response = requests.get(url)

if response.status_code == 200:
    print("Success!")
    if 'application/json' in response.headers.get('Content-Type', ''):  # Check if the response is JSON.'' means default value if key is not found.
                                                                        # It returns an empty string if the key is not found.
        data = response.json() # Parse the JSON response
        print(data) # Print the parsed JSON data

    print(response.text) # This will print the content of the response as a string
else:
    print("Enter while fetching", response.status_code) # Print an error message if the request failed
   
```

```python
import requests
url = 'https://httpbin.org/post'
data = {'username':'user', 'password':'pass'}
response = requests.post(url, data=data)

if response.status_code == 200:
    print("Login Successful!")
    print(response.text) # This will print the content of the response as a string
    print(response.url) # This will print the URL of the request
else:
    print("Login Failed. Status code:", response.status_code)
```

```python
import requests
url = 'https://jsonplaceholder.typicode.com/posts' # Example URL for a JSON API
params = {'userId': 1, 'id': 2} # Example query parameters
response = requests.get(url, params=params) # This will send a GET request with query parameters

#The actual URL will be 'https://jsonplaceholder.typicode.com/posts?userId=1&id=2'

print(response.url)  # This will print the full URL with query parameters
print(response.text)  # This will print the content of the response as a string
```

```python
import requests
url = 'https://httpbin.org/get'

username = 'user' #credentials for basic authentication
password = 'pass'

response = requests.get(url, auth=(username, password))  # This will send a GET request with basic authentication
if response.status_code == 200:   # Check if the request was successful
    print("Authentication Successful!")
    print(response.text)  # This will print the content of the response as a string
else:
    print(f'Authentication Failed. Status code: {response.status_code}')  # Print an error message if the request failed'
```

```python
from bs4 import BeautifulSoup

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie&quot; class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie&quot; class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie&quot; class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
"""


soup = BeautifulSoup(html_doc, 'html.parser') # This will parse the HTML document
print(soup.title)  # This will print the title tag
print(soup.p) # This will print the first paragraph tag
print(soup.a)  # This will print the first anchor tag

p_tag =soup.find('p')
print(p_tag)  # This will print the first paragraph tag found
```

```python
p_tags = soup.find_all('p') # This will print all anchor tags found
print(p_tags)  # This will print all paragraph tags found
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.example.com')

if response.status_code == 200:   
    print("Sucess")
    soup = BeautifulSoup(response.text, 'html.parser')
    main_header = soup.find('h1') 
    if main_header:
        print("Main Header:", main_header.text)
    else:
        print("No main header found")
    
else:
    print('Error fetching the HTML document')  

```
